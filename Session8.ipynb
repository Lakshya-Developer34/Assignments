```python
#Task 1
a = ["Zomato" , "Swiggy" , "DoorDash" , "UberEats" , "Deliveroo"]
for b in a:
    print(b)
```

```python
#Task 2
step_count = [2000,5000,11000,12000,8000,11000,15000]
# for i in step_count : 
#     if i > 10000 :
#         print(i)
#         break
# step_count = [2000, 5000, 10000, 12000, 8000, 11000, 15000]

i = 0

while i < len(step_count):
    if step_count[i] > 10000:
        print(step_count[i])
        break
    i += 1
```

```python
#Task 3
IPL_Team = ["Mumbai_Indians" , "Chennai_Super_Kings" , "RCB" , "SRH" , "Gujarat_Titans"]
for i in IPL_Team :
        if len(i) <= 6:
                continue
        print(i)
```

```python
#Task 4
songs_duration = [300 , 400 , 250 , 230 , 290]
song = 1
for i in songs_duration :
    print("Song", song ,":", i , "Seconds")
    song+=1
```

```python
#Task 5
item_prices = [200, 250, 0, 10, 600, 500, 300]

total = 0

for item_price in item_prices:

    if item_price == 0:
        continue

    if total + item_price > 2000:
        break

    total += item_price

print("Final Total =", total)
```
