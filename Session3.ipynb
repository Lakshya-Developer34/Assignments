Task 3. Create a script that asks the user to input their cricket score as a string, converts it to an int, and prints 'Half-century!' if the score is 50 or more, otherwise prints 'Keep going!'.<br><br><em><strong>Constraint:</strong> Use input(), int(), and if-else.</em>


```python
score=input()
output=int(score)
if output == 100:
    print("Century")
elif output > 100:
    print("More Than Century")
elif output == 50:
    print("Half-Century")
elif output > 50:
    print("More Than Half-Century")
else:
    print("Score is not more than 50")
```
