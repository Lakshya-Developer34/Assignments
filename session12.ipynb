Task 3. Download a sample CSV file of IPL match scores (you can create your own with columns: match_id, team1, team2, winner) and write a Python script to read the file and print the winner of each match using the csv module.

```python
ipl_score = [
    ["match_id" , "team1" , "team2" ,"winners"],
    [1,"CSK" , "MI" , "CSK"],
    [2,"KKR" , "MI" , "MI"],
    [3,"CSK" , "PBKS" , "CSK"],
    [4,"GT" , "RR" , "RR"]
]
```

```python
import csv
with open("match.csv", "w", newline="") as file3:
    write = csv.writer(file3)

    write.writerows(ipl_score)
```

```python
import csv
with open("match.csv" , "r") as file4:
    reader = csv.DictReader(file4)    # reader = csv.reader(file4)   
    for row in reader:
        print("winners :" , row["winners"])
```
