Task 3. Given a dictionary called food_prices with 5 Zomato food items as keys and their prices as values, write code to display all items that cost more than ₹200.


```python
food_prices = {"Burger" : 190 , "Pizza" : 220 , "Sandwich" : 80 , "Cold coffee" : 50 , "Biryani" : 330}
for item in food_prices:
    if food_prices[item] > 200:
        print(item, ":", food_prices[item])
```

Task 5. Write a function get_unique_artists(spotify_playlist1, spotify_playlist2) that takes two sets of artist names and returns a set of all unique artists across both playlists (set union).<br><br><em><strong>Hint:</strong> Use the union() method or the | operator for sets.</em>

```python
def get_unique_artists(spotify_playlist1, spotify_playlist2):
    # return spotify_playlist1.union(spotify_playlist2)  OR
    return spotify_playlist1 | spotify_playlist2

playlist1 = {"Arijit Singh", "Atif Aslam", "Shreya Ghoshal"}
playlist2 = {"Atif Aslam", "Neha Kakkar", "Armaan Malik"}

result = get_unique_artists(playlist1, playlist2)
print(result)
```
